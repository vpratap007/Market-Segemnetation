```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.utils import check_random_state, resample
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Set the random seed for reproducibility
random_state = check_random_state(1234)

# Example data: replace this with your actual data
MD_x = np.random.rand(100, 10)  # Assuming MD.x is a 100x10 array for example purposes

# Perform k-means clustering for a range of cluster numbers (2 to 8) with 10 repetitions each and bootstrapping
range_n_clusters = range(2, 9)
nrep = 10
nboot = 100
bootstrap_results = {n_clusters: [] for n_clusters in range_n_clusters}

for n_clusters in range_n_clusters:
    for _ in range(nrep):
        # Fit the k-means model
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
        kmeans.fit(MD_x)
        
        # Bootstrapping
        for _ in range(nboot):
            X_resampled = resample(MD_x, random_state=random_state)
            kmeans_boot = KMeans(n_clusters=n_clusters, random_state=random_state)
            kmeans_boot.fit(X_resampled)
            bootstrap_results[n_clusters].append(kmeans_boot.inertia_)

# Example of how to process or print the bootstrap results
for n_clusters, inertias in bootstrap_results.items():
    print(f"Number of clusters: {n_clusters}")
    print(f"Mean Inertia: {np.mean(inertias)}")
    print(f"Standard Deviation of Inertia: {np.std(inertias)}\n")

# You can further analyze the `bootstrap_results` dictionary to assess cluster stability
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import check_random_state, resample
from sklearn.metrics import adjusted_rand_score
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Set the random seed for reproducibility
random_state = check_random_state(1234)

# Example data: replace this with your actual data
MD_x = np.random.rand(100, 10)  # Assuming MD.x is a 100x10 array for example purposes

# Perform k-means clustering for a range of cluster numbers (2 to 8) with 10 repetitions each and bootstrapping
range_n_clusters = range(2, 9)
nrep = 10
nboot = 100
bootstrap_ari = {n_clusters: [] for n_clusters in range_n_clusters}

for n_clusters in range_n_clusters:
    for _ in range(nrep):
        # Fit the k-means model to the original data
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
        kmeans.fit(MD_x)
        original_labels = kmeans.labels_
        
        # Bootstrapping
        for _ in range(nboot):
            X_resampled = resample(MD_x, random_state=random_state)
            kmeans_boot = KMeans(n_clusters=n_clusters, random_state=random_state)
            kmeans_boot.fit(X_resampled)
            boot_labels = kmeans_boot.labels_
            ari = adjusted_rand_score(original_labels, boot_labels)
            bootstrap_ari[n_clusters].append(ari)

# Create a boxplot of the ARI values for different numbers of clusters
plt.figure(figsize=(10, 6))
plt.boxplot([bootstrap_ari[n_clusters] for n_clusters in range_n_clusters], labels=list(range_n_clusters))
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Global Stability Boxplot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.utils import check_random_state

# Set the random seed for reproducibility
random_state = check_random_state(1234)

# Example data: replace this with your actual data
MD_x = np.random.rand(100, 10)  # Assuming MD.x is a 100x10 array for example purposes

# Perform Gaussian mixture modeling using scikit-learn for a range of components (clusters)
range_n_components = range(2, 9)
nrep = 10
gmm_models = {}

for n_components in range_n_components:
    gmm_models[n_components] = []
    for _ in range(nrep):
        gmm = GaussianMixture(n_components=n_components, random_state=random_state)
        gmm.fit(MD_x)
        gmm_models[n_components].append(gmm)

# Print the log-likelihood, AIC, BIC, and ICL values for each model
print("Model Selection Results:")
print("iter converged k k0 logLik AIC BIC ICL")
for n_components, models in gmm_models.items():
    for idx, model in enumerate(models):
        loglik = model.lower_bound_  # Log-likelihood
        aic = model.aic(MD_x)  # AIC
        bic = model.bic(MD_x)  # BIC
        # ICL (Integrated Complete-data Likelihood)
        icl = loglik - (0.5 * model.n_components) * np.log(MD_x.shape[0])
        print(f"{idx + 2} {model.converged_} {n_components} {n_components} {loglik:.3f} {aic:.2f} {bic:.2f} {icl:.2f}")
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.utils import check_random_state
from sklearn.metrics import confusion_matrix
import pandas as pd

# Set the random seed for reproducibility
random_state = check_random_state(1234)

# Example data: replace this with your actual data
MD_x = np.random.rand(1000, 10)  # Assuming MD.x is a 1000x10 array for example purposes

# Perform k-means clustering for 4 clusters
kmeans = KMeans(n_clusters=4, random_state=random_state)
kmeans.fit(MD_x)
kmeans_clusters = kmeans.predict(MD_x)

# Fit Gaussian Mixture Model using the clusters from k-means as initialization
gmm_init = GaussianMixture(n_components=4, random_state=random_state, init_params='kmeans')
gmm_init.fit(MD_x)

# Log-likelihood of the GMM fitted with k-means initialization
log_likelihood_init = gmm_init.score(MD_x) * MD_x.shape[0]

# Fit Gaussian Mixture Model using k-means clusters directly as cluster labels
gmm_direct = GaussianMixture(n_components=4, random_state=random_state)
gmm_direct.fit(MD_x, kmeans_clusters)

# Log-likelihood of the GMM fitted with k-means clusters directly as labels
log_likelihood_direct = gmm_direct.score(MD_x) * MD_x.shape[0]

# Predict clusters using the Gaussian Mixture Model
gmm_clusters = gmm_init.predict(MD_x)

# Generate a cross-tabulation (contingency table) comparing the clusters
contingency_table = confusion_matrix(kmeans_clusters, gmm_clusters)

# Convert the contingency table to a pandas DataFrame for better visualization
contingency_df = pd.DataFrame(contingency_table, 
                              index=[f"kmeans {i}" for i in range(1, 5)], 
                              columns=[f"mixture {i}" for i in range(1, 5)])

# Print the cross-tabulation table
print("Cross-tabulation of k-means and mixture model clusters:")
print(contingency_df)

# Print log-likelihood values
print(f"Log-likelihood of GMM with k-means initialization: {log_likelihood_init}")
print(f"Log-likelihood of GMM with k-means clusters as labels: {log_likelihood_direct}")
```

```python
from statsmodels.graphics.mosaicplot import mosaic
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Assuming md is your actual DataFrame

# Function to label encode categorical variables
def labelling(x):
    md[x] = LabelEncoder().fit_transform(md[x])
    return md

# Columns to label encode
cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

# Apply label encoding
for i in cat:
    labelling(i)

# Selecting the 11 columns
df_eleven = md.loc[:, cat]

# Perform KMeans clustering
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df_eleven)
md['cluster_num'] = kmeans.labels_

# Create a contingency table
crosstab = pd.crosstab(md['cluster_num'], md['Like'])

# Print the actual columns in crosstab to debug
print("Actual columns in crosstab:", crosstab.columns.tolist())

# Ensure the 'Like' column categories are in the correct order
like_order = ['I hate it!-5', '-4', '-3', '-2', '-1', '0', '1', '2', '3', '4', 'I love it!+5']
# Adjust the like_order based on actual columns present in crosstab
like_order = [col for col in like_order if col in crosstab.columns]

print("Adjusted like_order:", like_order)

crosstab = crosstab[like_order]

# Plotting the mosaic plot
plt.rcParams['figure.figsize'] = (7, 5)
mosaic(crosstab.stack())
plt.show()
```
